```python
import scipy.special as sp

def dgp5(E, gt=10, beta=5, gamma=1.0):
    X = np.random.normal(size=(E.shape[0], 1))

    piZ = sp.expit((np.mean(X, axis=1) + gamma * np.mean(np.matmul(E, X),
                                                         axis=1)) / 10)  # + np.random.normal(0, 1, size=X.shape[0]))
    Z = np.random.binomial(1, piZ)

    deg = E.sum(axis=1)
    # peer_effect = (np.matmul(E, Z) > deg/2).astype(int)
    # peer_effect = np.matmul(E, Z) / deg
    peer_effect = np.matmul(E, Z)

    Y = Z * gt + peer_effect * beta + np.sum(X, axis=1) + np.sum(np.matmul(E, X), axis=1)

    df = pd.DataFrame(X, columns=['X%d' % (i) for i in range(X.shape[1])],
                      index=['u%d' % (i) for i in range(X.shape[0])])
    df['Y'] = Y
    df['Z'] = Z
    network = pd.DataFrame(E, columns=['u%d' % (i) for i in range(X.shape[0])],
                           index=['u%d' % (i) for i in range(X.shape[0])])
    df_true = pd.DataFrame()
    df_true['piZ'] = piZ
    return df, network, df_true, X

with open(f'/Users/asus/Desktop/MyResearch/baselines/Causal_Relational/Dataset/Simulated_SBM/1000_network.pkl', 'rb') as f:
    E = pkl.load(f)
dgp5(E)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

data = pd.read_csv("../results/intermediate.csv")

# Extract the relevant columns
X = data[['resT', 'resPE']].values
y = data['resY'].values
T = data['T'].values

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Define a function to plot the 3D scatter plot and the regression hyperplane
def plot_3d_scatter_with_hyperplane(X, y, T, model):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Scatter plot with colored points based on 'T'
    colors = np.where(T == 0, 'black', 'green')
    scatter = ax.scatter(X[:, 0], X[:, 1], y, c=colors, marker='.')

    # Create a meshgrid for the hyperplane
    x_surf, y_surf = np.meshgrid(np.linspace(X[:, 0].min(), X[:, 0].max(), 100), 
                                 np.linspace(X[:, 1].min(), X[:, 1].max(), 100))
    z_surf = model.predict(np.c_[x_surf.ravel(), y_surf.ravel()]).reshape(x_surf.shape)

    # Plot the hyperplane
    ax.plot_surface(x_surf, y_surf, z_surf, color='y', alpha=0.5)

    # Set labels with increased fontsize
#     ax.set_xlabel('resT', fontsize=35, labelpad=20)
#     ax.set_ylabel('resPE', fontsize=35, labelpad=16)
#     ax.set_zlabel('resY', fontsize=35, labelpad=17)  # Adjust labelpad for z-axis

    # Create a custom legend
    handles = [
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='black', markersize=22, label='Control'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green', markersize=22, label='Treated')
    ]
    ax.legend(bbox_to_anchor=(1.05, 1), handles=handles, loc='best', fontsize=28, borderaxespad=0.)

    ax.view_init(elev=15, azim=-30)

#     plt.title('Final regression of resY on resT and resPE', fontsize=34)
    plt.savefig('../results/regression')
    plt.show()

# Call the function to plot
plot_3d_scatter_with_hyperplane(X, y, T, model)
```

```python
import numpy as np

def build_adjacency_matrix(file_path):
    # Read the file and extract the edges
    with open(file_path, 'r') as file:
        edges = [tuple(map(int, line.strip().split())) for line in file]

    # Extract all unique node indexes and map them to a contiguous range
    unique_nodes = set()
    for edge in edges:
        unique_nodes.update(edge)
    
    node_index_map = {node: idx for idx, node in enumerate(unique_nodes)}
    num_nodes = len(unique_nodes)

    # Initialize the adjacency matrix with zeros
    adjacency_matrix = np.zeros((num_nodes, num_nodes), dtype=int)

    # Populate the adjacency matrix
    for edge in edges:
        i, j = node_index_map[edge[0]], node_index_map[edge[1]]
        adjacency_matrix[i][j] = 1
        adjacency_matrix[j][i] = 1  # Assuming an undirected graph

    return adjacency_matrix, node_index_map

# Example usage
file_path = '/Users/asus/Downloads/flickrEdges (1).txt'
adjacency_matrix, node_index_map = build_adjacency_matrix(file_path)
print("Adjacency Matrix:")
print(adjacency_matrix)
print("Node Index Mapping:")
print(node_index_map)
```
